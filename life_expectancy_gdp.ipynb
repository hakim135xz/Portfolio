```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Importing the necessary modules

```python
# reading the csv file, convert to df

all_data = pd.read_csv('C:/Users/hakim/Documents/Codecademy_portfolio_projects/Life-Expectancy-and-GDP-Starter/all_data.csv')
```

```python
# changing the column name
all_data.columns = ['country', 'year', 'life_expectancy', 'gdp']
```

```python
# life expectancy increases over the years for some countries
# mexico df
mexico_data = all_data[all_data.country == 'Mexico']
# chile
chile_data = all_data[all_data.country == 'Chile']
# Us
us_data = all_data[all_data.country == 'United States of America']
# germany
german_data = all_data[all_data.country == 'Germany']
```

```python
# plot starts here
# mexico
sns.scatterplot(data=mexico_data, x=mexico_data.year, y=mexico_data.life_expectancy)
plt.show()
plt.clf()
```

```python
# plot starts here
# chile
sns.scatterplot(data=chile_data, x=chile_data.year, y=chile_data.life_expectancy)
plt.show()
plt.clf()
```

```python
# plot starts here
# usa
sns.scatterplot(data=us_data, x=us_data.year, y=us_data.life_expectancy)
plt.show()
plt.clf()
```

```python
# plot starts here
# germany
sns.scatterplot(data=german_data, x=german_data.year, y=german_data.life_expectancy)
plt.show()
plt.clf()
```

Analysing GDP

```python
# plotting boxplot
sns.boxplot(data=all_data, x=all_data.year, y=all_data.gdp, hue=all_data.country)
plt.show()
plt.clf()
```
